```python
# Importing dependencies
import requests
import pandas as pd
import matplotlib as plt
from scipy.stats import f_oneway

# Load data from CSV file
resilience_texas = pd.read_csv("Resilience Resources/Climate_Resilience_Screening_Index_Texas.csv")
resilience_north_carolina = pd.read_csv("Resilience Resources/Climate_Resilience_Screening_Index_North_Carolina.csv")
resilience_new_york = pd.read_csv("Resilience Resources/Climate_Resilience_Screening_Index_New_York.csv")

# Calculate the statics for each state

# Min, Max, Mean, Standard Deviation, Variance for Texas

min_texas = resilience_texas['CRSI'].min()
max_texas = resilience_texas['CRSI'].max()
mean_texas = resilience_texas['CRSI'].mean()
stdev_texas = resilience_texas['CRSI'].std()
variance_texas = resilience_texas['CRSI'].var()

# Print statistics result
print("Statistics for Texas:")
print("Minimum CRSI:", min_texas)
print("Maximum CRSI:", max_texas)
print("Mean CRSI:", mean_texas)
print("Standard Deviation CRSI:", stdev_texas)
print("Variance CRSI:", variance_texas)



# Min, Max, Mean, Standard Deviation, Variance for North Carolina

min_nc = resilience_north_carolina['CRSI'].min()
max_nc = resilience_north_carolina['CRSI'].max()
mean_nc = resilience_north_carolina['CRSI'].mean()
stdev_nc = resilience_north_carolina['CRSI'].std()
variance_nc = resilience_north_carolina['CRSI'].var()

# Print statistics result
print("Statistics for North Carolina:")
print("Minimum CRSI:", min_north_carolina)
print("Maximum CRSI:", max_north_carolina)
print("Mean CRSI:", mean_north_carolina)
print("Standard Deviation CRSI:", stdev_north_carolina)
print("Variance CRSI:", variance_north_carolina)


# Min, Max, Mean, Standard Deviation, Variance for New York

min_ny = resilience_new_york['CRSI'].min()
max_ny = resilience_new_york['CRSI'].max()
mean_ny = resilience_new_york['CRSI'].mean()
stdev_ny = resilience_new_york['CRSI'].std()
variance_ny = resilience_new_york['CRSI'].var()

# Print statistics result
print("Statistics for New York:")
print("Minimum CRSI:", min_new_york)
print("Maximum CRSI:", max_new_york)
print("Mean CRSI:", mean_new_york)
print("Standard Deviation CRSI:", stdev_new_york)
print("Variance CRSI:", variance_new_york)


# Perform an ANOVA test to assess the statistical significance differences between the states.

f_statistic, p_value = f_oneway(resilience_texas['CRSI'], resilience_north_carolina['CRSI'], resilience_new_york['CRSI'])

if p_value < 0.05:
    print("The differences between the states are statistically significant.")
else:
    print("The differences between the states are not statistically significant.")


# Establish Visualization statistics for state Climate Resilience Index (CRSI)

# Bar chart for mean CRSI values
mean_values = [mean_texas, mean_north_carolina, mean_new_york]
states = ['Texas', 'North Carolina', 'New York']

plt.figure(figsize=(8, 6))
plt.bar(states, mean_values)
plt.xlabel('State')
plt.ylabel('Mean CRSI')
plt.title('Mean Climate Resilience Screening Index for Each State')
plt.show()

# Box plot for CRSI distribution
plt.figure(figsize=(8, 6))
plt.boxplot([resilience_texas['CRSI'], resilience_north_carolina['CRSI'], resilience_new_york['CRSI']], labels=states)
plt.xlabel('State')
plt.ylabel('CRSI')
plt.title('Climate Resilience Screening Index Distribution by State')
plt.show()

# ANOVA plot for CRSI analysis
plt.figure(figsize=(8, 6))
plt.bar(['F-Statistic', 'P-Value'], [f_statistic, p_value])
plt.xlabel('ANOVA Results')
plt.ylabel('Value')
plt.title('ANOVA Results for CRSI Comparison')
plt.show()



```
